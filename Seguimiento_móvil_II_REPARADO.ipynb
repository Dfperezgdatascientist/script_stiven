```python
seg_x_concat=pd.read_csv("Cobranzas_movil_(Despliegue)/asesores_sin_concatenar.csv",sep=';',encoding='utf-8')
file_paths3 = "Cobranzas_movil_(Despliegue)/Download_folder_Adminfo/{}_seg.zip".format(yesterday)
cc=['Identificación', 'Referencia Pago', 'Fecha De Gestión',
'Clase de Gestión', 'Ciclo', 'Nota', 'Teléfono',
'Descripcion Causal Mora', 'Regional', 'Departamento',
'Marca Portafolio', 'Codigo Traslado', 'Duración',
'Descripcion Mejor Codigo Gestion', 'Causal Mora', 'Fecha Churn',
'Fecha De carga', 'Tipo Inventario', 'Tipo Plan', 'Estado De Cuenta',
'Plan', 'Estado Plan', 'Ciudad', 'Castigada', 'Acción', 'Días Mora',
'Etapa Campaña Externa', 'Antiguedad Producto', 'Fecha De Compromiso',
'Inmunidad B2B', 'Valor De Compromiso', 'Canal B2B',
'Calificación Habito Pago', 'Edad Mora', 'Estado Localización',
'Fecha de suspensión decreto']
def open_files_seg(paths):
    

    data = []
    for path in paths:
        zf = zipfile.ZipFile(path)#Función de apertura de .zip
        data.append(pd.read_csv(#Función de lectura paralelizada
                    zf.open(zf.filelist[0].filename),#Abre el primer archivo del .zip
                    sep = "|",
                    encoding = "utf-8",
                    dtype=str,
                    error_bad_lines=False,
                    usecols=cc
                    )
                )
        print(path,"ok")
    return pd.concat(data,ignore_index = True)

seg1 = open_files_seg([file_paths3])
seg1=seg1[~seg1.duplicated(keep='last')]
seg1.loc[seg1['Identificación'].str.len() < 7, "Identificación"] = np.nan  
seg1=seg1.dropna(subset=['Identificación'])
seg=pd.concat([seg1,seg_x_concat])
seg.to_csv("Cobranzas_movil_(Despliegue)/asesores_sin_concatenar.csv",sep=';',encoding='utf-8',index=False)
seg['Descripcion Mejor Codigo Gestion']=seg['Descripcion Mejor Codigo Gestion'].str.upper().str.strip()
seg1=seg1[~seg1.duplicated(keep='last')]
seg1.loc[seg1['Identificación'].str.len() < 7, "Identificación"] = np.nan  
seg1=seg1.dropna(subset=['Identificación'])
seg=pd.concat([seg1,seg_x_concat])
seg.to_csv("Cobranzas_movil_(Despliegue)/asesores_sin_concatenar.csv",sep=';',encoding='utf-8',index=False)
seg['Descripcion Mejor Codigo Gestion']=seg['Descripcion Mejor Codigo Gestion'].str.upper().str.strip()
seg['Descripcion Mejor Codigo Gestion']=seg['Descripcion Mejor Codigo Gestion'].str.upper().str.strip()
diicionario_cambios={'YA PAGO':'PAGO REALIZADO','BUZON DE VOZ':'SE DEJO MENSAJE EN BUZON',
                    'FUERA DE SERVICIO/INVALIDO/ERRADO':'DANADO - FUERA DE SERVICIO','DEUDOR ILOCALIZADO':'ILOCALIZADO',
                    'MENSAJE CON TERCEROS CASA':'SE DEJO MENSAJE CON TERCERO','NO HUBO ACUERDO (GESTIONABLE)':'NO HUBO ACUERDO'}
seg['Descripcion Mejor Codigo Gestion']=seg['Descripcion Mejor Codigo Gestion'].replace(diicionario_cambios)
seg['Descripcion Mejor Codigo Gestion'][seg['Descripcion Mejor Codigo Gestion']=='']='SIN CODIGO'
seg['Fecha De Gestión']=seg.apply(
lambda row: row['Fecha De Gestión'][:-3]  if (row['Fecha De Gestión'].endswith("-05")) else (row['Fecha De Gestión']),axis=1)
seg['Fecha De Gestión']=pd.to_datetime(seg['Fecha De Gestión'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
seg['Ano_seg']=seg['Fecha De Gestión'].dt.year
seg['mes_seg']=seg['Fecha De Gestión'].dt.month
seg['hora_seg']=seg['Fecha De Gestión'].dt.hour
seg['Ano_mes']=seg['Ano_seg'].astype('str')+"_"+seg['mes_seg'].astype('str')
seg.loc[seg['Identificación'].str.len() < 7, "Identificación"] = np.nan  
seg=seg.dropna(subset=['Identificación'])
cant_mesco=seg.groupby(['Identificación'])['Ano_mes'].nunique().reset_index().rename(columns={'Ano_mes':'Cantidad_de_cobros_diff_mes'})
cant_rec=seg.groupby(['Identificación','Ano_seg','mes_seg']).size().reset_index(name='Cantidad_recontactos_mes').sort_values(['Identificación','Ano_seg','mes_seg'],ascending=[True,True,True]).drop_duplicates('Identificación',keep='last') 
seg=pd.merge(seg,cant_mesco,on='Identificación').merge(cant_rec[['Identificación','Cantidad_recontactos_mes']],on='Identificación')
seg['class_Cantidad_de_cobros_diff_mes']=\
        pd.cut(x=seg['Cantidad_de_cobros_diff_mes'],
            bins=[0,1,2,3,6,10,17,1000],
            labels=["1","2","3","4-6","7-10", "11-17", ">17"])
seg['class_Cantidad_recontactos_mes']=\
                pd.cut(x=seg['Cantidad_recontactos_mes'], 
                    bins=[0,1,2,3,6,10,17,1000], 
                    labels=["1","2","3","4-6","7-10", "11-17", ">17"])
conteo3=seg['Descripcion Mejor Codigo Gestion'].value_counts().iloc[:10].index.tolist()
seg['Descripcion Mejor Codigo Gestion_2']=seg.apply(
lambda row: row['Descripcion Mejor Codigo Gestion']  if (row['Descripcion Mejor Codigo Gestion'] in conteo3)
else 'OTRA GESTION',axis=1)
seg_2=pd.concat([seg,pd.get_dummies(seg['Descripcion Mejor Codigo Gestion_2'])],axis=1)
grupo=seg_2.groupby(['Ano_seg','mes_seg','Identificación']).agg({'Descripcion Mejor Codigo Gestion_2':'np.array',
                    'Descripcion Mejor Codigo Gestion_2':'sum',
                   # 'PROMESA DE PAGO':'sum',
                    'PAGO REALIZADO':'sum',
                    'SE DEJO MENSAJE EN BUZON':'sum',
                    'NO CONTESTA / OCUPADO':'sum',
                    'OTRA GESTION':'sum',
                    'NO HUBO ACUERDO':'sum',
                    'SE DEJO MENSAJE CON TERCERO':'sum',
                    # 'DEUDOR SOLICITAR CUPON DE PAGO':'sum',
                    'DEUDOR REALIZA PROMESA DE PAGO TOTAL':'sum',                                                
                                        }).rename(columns={'OTRA GESTION':'OTRA GESTION_sum'}).reset_index()
segui=pd.merge(seg_2.drop(conteo3,axis=1),grupo.drop(['Descripcion Mejor Codigo Gestion_2'],axis=1),
left_on=['Ano_seg','mes_seg','Identificación'],
right_on=['Ano_seg','mes_seg','Identificación'])
segui['Antiguedad Producto']=segui['Antiguedad Producto'].str.replace('En MaduraciÃ³n','En Maduración').str.replace('AntiguÃ³','Antiguó')
segui.drop(['Descripcion Mejor Codigo Gestion_2'],axis=1,inplace=True)
segui.to_csv('Cobranzas_movil_(Despliegue)/asesores_concatenada.csv',index=False,sep=';',encoding='utf-8')
print(' Seguimiento concatenado y copiado con exito')
```
