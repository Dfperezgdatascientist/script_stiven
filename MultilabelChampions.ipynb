```
import pandas as pd
df = pd.read_excel("FallasFuncionamiento.xlsx")
df.head()
```

```
#Limpieza
import re
# Limpieza, carecteres especiales y signos de puntuacion
def limpieza_basica(body):
    body = body.lower() #todo en minuscula
    body = re.sub(" \d+",' ', body) #eliminar correos
    body = re.sub("\S+@\S+",' ', body) #Eliminar correros
    body = re.sub('\w*[0-9]\w*',' ',body) #Remover codigos letras y numeros
    body = re.sub("[0-9]{2}[\/,:][0-9]{2}[\/,:][0-9]{2,4}",'',body) #Quitar fecha y hora
    #body = re.sub('[\d+]',' ', body) #quitar numeros dentro de palabras
    body = re.sub('[#*>=\\!-]','',body) # Eliminar caracteres 
    body = re.sub('/ /','',body) 
    #body = re.sub(r'[^\w\s]','',body) #eliminar signos de puntuacion
    #body = re.sub('[%s]' %re.escape(string.punctuation),' ', body) # Eliminar signos puntuacion
    body= re.sub(r"[\r\n]+",' ', body) # saltos de linea iguales
    body=re.sub("plu" , " ", body)
    body=re.sub(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+",' ', body)
    body=re.sub(" p. m." , " ", body)  
    body=re.sub("cliente se comunica" , "", body)  
    body=re.sub("el cliente se comunica" , "", body)  
    body=re.sub("\s\s+" , " ", body) #quita varios espacios
    body = re.sub("\S+@\S+",' ', body) #Eliminar correros
    return body
limpiar = lambda x: limpieza_basica(x)
```

```
df["Des"] = df["Descripcion"].apply(limpiar)
```

```
# dropping ALL duplicte values 
df.drop_duplicates(subset ="Caso", 
                     keep = False, inplace = True)
```

```
from simpletransformers.classification import (
    MultiLabelClassificationModel, MultiLabelClassificationArgs
)
import pandas as pd
import logging
```

```
import pandas as pd
df = pd.read_excel("PruebaIII.xlsx")
df.head()
```

```
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```
df['labels'] = list(zip(df.DificultadEncendido.tolist(), df.Bateria.tolist(),  df.Sonido.tolist(),  df.Pantalla.tolist()))
df.head()
```

```
df['text'] = df['Des']
df.head()
```

```
col = ['text', 'labels']
df = df[col]
df.head()
```

```
from sklearn.model_selection import train_test_split
train_df, eval_df = train_test_split(df, test_size=0.2)
```

```
# Optional model configuration
model_args = MultiLabelClassificationArgs(num_train_epochs=3)
```

```
from simpletransformers.classification import MultiLabelClassificationModel

#'roberta', 'chriskhanhtran/spanberta'
#model = MultiLabelClassificationModel('bert', 'dccuchile/bert-base-spanish-wwm-cased', num_labels=4)
model = MultiLabelClassificationModel('roberta', 'chriskhanhtran/spanberta',
                                      num_labels=4)
```

```
# Train the model
model.train_model(train_df)
```

```
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(
    eval_df
)
```
