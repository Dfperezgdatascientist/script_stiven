```python
path=r'D:\3. Random_scripts\Haceb\Calidad'
filenames = glob.glob(path+"/*.xlsx")
filenames
```

```python
df = pd.DataFrame()
for file in range(len(filenames)):
    dfini = pd.read_excel(filenames[file])
    dfini['filename'] = filenames[file].split('\\')[-1]
    df = df.append(dfini)
df
```
